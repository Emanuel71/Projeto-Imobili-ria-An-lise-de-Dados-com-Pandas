# Tratando e filtrando os dados

## Lidando com dados nulos

```python
# Verificando quais colunas, tem valores nulos
df.isnull().sum()
```

```python
# Alterando os valotes null por 0.
df = df.fillna(0)
df
```

```python
df.isnull().sum()
```

## Removendo registros

```python
# Selecionando registro que tenha valores 0, da coluna Valor e Condominio.
regisitro_a_remover = df.query('Valor == 0 | Condominio == 0').index
regisitro_a_remover
```

```python
# Removendo registro 
df.drop(regisitro_a_remover, axis=0, inplace=True)
```

```python
df.query('Valor == 0 | Condominio == 0')
```

```python
# Iremos remover a coluna Tipo, por contem um registro unique
df.drop('Tipo', axis=1, inplace=True)
```

```python
df.head()
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



```python
# Forma 1 - manual
selecao1 = (df['Quartos'] == 1) & (df['Valor'] < 1200)

# Atribuindo esse filtro ao DateFrame
df_1 = df[selecao1] 
df_1
```

```python
# Forma 2 - eficiente
df.query('Quartos == 1 & Valor < 1200')
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
# Forma 1 - manual 
selecao2 = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
df_2 = df[selecao2]
df_2
```

```python
# Forma 2 - eficiente 
df.query('Quartos >= 2 & Valor < 3000 & Area > 70')
```

# Manipulando os dados

## Criando colunas numéricas

```python
# Fazendo a importanção dos dados originais
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
# Criando uma coluna de valor mensal gasto por imoveis
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
# Criando uma coluna de valor anual gasto por imoveis
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
# astype = usamos para converter colunas numericas em str, para podemos fazer a concatenação
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                    dados['Quartos'].astype(str) + ' quarto(s) ' + \
                    ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
# Criandp uma coluna para informar se temos suite ou não
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim'  if x > 0 else 'Não')
dados.head()
```

```python
dados.to_csv('dados_completo_dev.csv', index=False, sep= ';')
```
